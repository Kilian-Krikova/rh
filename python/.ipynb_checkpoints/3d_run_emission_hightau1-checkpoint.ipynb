## cb24bih_s385_x80_y102 with width 20

```python
rh3d_crd_width20 = rhanalyze.rhout('runs/high_tau1/cb24bih_x80_y102/cb24bih_s385_x80_y102_width20_CaH_CRD_Heps_rh15d')
```

```python
x_cen = 80
y_cen = 102
width = 20
height = 20
inten=data.ray.intensity[x_cen-width:x_cen+width+1, y_cen-height:y_cen+height+1,:].data
inten_hnei=data_hnei.ray.intensity[x_cen-width:x_cen+width+1, y_cen-height:y_cen+height+1,:].data
```

```python
wvl_search= 397.1202
wvl_ind_15d = np.argmin(np.abs(data.ray.wavelength.data - wvl_search))
wvl_ind_3d = np.argmin(np.abs(rh3d_crd_width20.spectrum.waves- wvl_search))
x_min = 60
y_min = 82
x_loc = [20,19,19,20]
y_loc = [18,20,23,20]
x_loc_rh15d = [x_min+x_loc[0], x_min+x_loc[1], x_min+x_loc[2], x_min+x_loc[3]]
y_loc_rh15d = [y_min+y_loc[0], y_min+y_loc[1], y_min+y_loc[2], y_min+y_loc[3]]

#%matplotlib widget
plt.figure(figsize=[12,7])
plt.subplot(2,3,1)
plt.imshow(inten[:,:,wvl_ind_15d].T, vmin=np.percentile(inten[:,:,wvl_ind_15d],1), vmax=np.percentile(inten[:,:,wvl_ind_15d],99))
plt.title('1.5D SE')
plt.scatter(x_loc, y_loc, s=30, marker='x',c='red', lw=2)

plt.subplot(2,3,4)
plt.imshow(rh3d_crd_width20.rays[0].I[:,:,wvl_ind_3d].T, vmin=np.percentile(inten[:,:,wvl_ind_15d],1), vmax=np.percentile(inten[:,:,wvl_ind_15d],99))
plt.title('3D SE')
plt.scatter(x_loc, y_loc, s=30, marker='x',c='red', lw=2)

plt.subplot(2,3,2)
plt.plot(rh3d_crd_width20.spectrum.waves, rh3d_crd_width20.rays[0].I[x_loc[0],y_loc[0],:], linewidth=2.0, label='SE 3D CRD width20')
#plt.plot(rh3d_prd_plane.spectrum.waves, rh3d_prd_plane.rays[0].I[0,0,:], linewidth=2.0, label='SE 3D PRD plane', ls='--')
plt.plot(data.ray.wavelength, data.ray.intensity[x_loc_rh15d[0], y_loc_rh15d[0],:], linewidth=2.0, label='SE 1.5D PRD', ls='--', c='cyan')
plt.title('x20 y18')
plt.legend()
plt.xlim([396.8, 397.34])
plt.ylim([0.01e-8, 1.2e-8])

plt.subplot(2,3,3)
plt.plot(rh3d_crd_width20.spectrum.waves, rh3d_crd_width20.rays[0].I[x_loc[1],y_loc[1],:], linewidth=2.0, label='SE 3D CRD width20')
plt.plot(data.ray.wavelength, data.ray.intensity[x_loc_rh15d[1], y_loc_rh15d[1],:], linewidth=2.0, label='SE 1.5D PRD', ls='--', c='cyan')
plt.title('x19 y20')
plt.legend()
plt.xlim([396.8, 397.34])
plt.ylim([0.01e-8, 1.5e-8])

plt.subplot(2,3,5)
plt.plot(rh3d_crd_width20.spectrum.waves, rh3d_crd_width20.rays[0].I[x_loc[2],y_loc[2],:], linewidth=2.0, label='SE 3D CRD width20')
plt.plot(data.ray.wavelength, data.ray.intensity[x_loc_rh15d[2], y_loc_rh15d[2],:], linewidth=2.0, label='SE 1.5D PRD', ls='--', c='cyan')
plt.title('x19 y23')
plt.legend()
plt.xlim([396.8, 397.34])
plt.ylim([0.01e-8, 2.5e-8])

plt.subplot(2,3,6)
plt.plot(rh3d_crd_width20.spectrum.waves, rh3d_crd_width20.rays[0].I[x_loc[3],y_loc[3],:], linewidth=2.0, label='SE 3D CRD width20')
plt.plot(data.ray.wavelength, data.ray.intensity[x_loc_rh15d[3], y_loc_rh15d[3],:], linewidth=2.0, label='SE 1.5D PRD', ls='--', c='cyan')
plt.title('x20 y20')
plt.legend()
plt.xlim([396.8, 397.34])
plt.ylim([0.01e-8, 1.0e-8])

plt.tight_layout()
```
